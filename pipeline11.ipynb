```python
def reduce_mem_usage(df):
    """ iterate through all the columns of a dataframe and modify the data type
        to reduce memory usage.
    """
    start_mem = df.memory_usage().sum() / 1024**2
    
    for col in df.columns:
        col_type = df[col].dtype
        
        if col_type != object:
            c_min = df[col].min()
            c_max = df[col].max()
            if str(col_type)[:3] == 'int':
                if c_min > np.iinfo(np.int8).min and c_max < np.iinfo(np.int8).max:
                    df[col] = df[col].astype(np.int8)
                elif c_min > np.iinfo(np.int16).min and c_max < np.iinfo(np.int16).max:
                    df[col] = df[col].astype(np.int16)
                elif c_min > np.iinfo(np.int32).min and c_max < np.iinfo(np.int32).max:
                    df[col] = df[col].astype(np.int32)
                elif c_min > np.iinfo(np.int64).min and c_max < np.iinfo(np.int64).max:
                    df[col] = df[col].astype(np.int64)
            else:
                #if c_min > np.finfo(np.float16).min and c_max < np.finfo(np.float16).max:
                #    df[col] = df[col].astype(np.float16)
                #el
                if c_min > np.finfo(np.float32).min and c_max < np.finfo(np.float32).max:
                    df[col] = df[col].astype(np.float32)
                else:
                    df[col] = df[col].astype(np.float64)
        #else:
            #df[col] = df[col].astype('category')

    end_mem = df.memory_usage().sum() / 1024**2
    print('Memory usage of dataframe is {:.2f} MB --> {:.2f} MB (Decreased by {:.1f}%)'.format(
        start_mem, end_mem, 100 * (start_mem - end_mem) / start_mem))
    return df
```

```python
data=reduce_mem_usage(df)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.decomposition import PCA
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
```

```python
array =data.values
array1 = y.values
X = array[:,0:7]
Y = array1[:,0]
seed = 7
Y
```

```python
from sklearn.preprocessing import normalize
X=normalize(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,Y, test_size = 0.3,random_state=1000)
```

```python
#Logistic regression pipeline
LogisticRegressionPipeline=Pipeline([('myscaler',MinMaxScaler()),('mypca',PCA(n_components=3)),('logistic_classifier',LogisticRegression())])
```

```python
DecisionTreePipeline=Pipeline([('myscaler',MinMaxScaler()),('mypca',PCA(n_components=3)),('decisiontree_classifier',DecisionTreeClassifier())])
```

```python
RandomForestPipeline=Pipeline([('myscaler',MinMaxScaler()),('mypca',PCA(n_components=3)),('randomforest_classifier',RandomForestClassifier())])
```

```python
#model training and validation
mypipeline=[LogisticRegressionPipeline, DecisionTreePipeline,RandomForestPipeline]
```

```python
accuracy=0.0
classifier=0
pipeline=" "
```

```python
# creating dictionary pipelines and training models
PipeLineDict={0:'LogisticRegression',1:'DecisionTree',2:'RandomForest'}
#fit the pipeline
for mypipe in mypipeline:
    mypipe.fit(X_train,y_train)
```

```python
for i,model in enumerate(mypipeline):
    print("{} TestAccuracy:{}".format(PipelineDict[i],model.score(X_test,y_test)))
```

```python
for i,model in enumerate(mypipeline):
    if model.score(X_test,y_test)>accuracy:
        accuracy=model.score(X_test,y_test)
        pipeline=model
        classifier=i
print('classifier with the nest accuracy:{}'.format(PipelineDict[classifier]))
```
